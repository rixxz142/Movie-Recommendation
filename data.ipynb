```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-Processing

```python
#load the data from the csv file 
movies = pd.read_csv('E:\MovieRecommendation\movies.csv')
movies.head()
```

```python
movies.shape
```

```python
#select the relevant features for recommendation
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
#replacing the null values with null string

for feature in selected_features:
    movies[feature] = movies[feature].fillna('') 
```

```python
#combine all 5 selected features
combine = movies['genres']+''+movies['keywords']+''+movies['tagline']+''+movies['cast']+''+movies['director']
```

```python
print(combine)
```

```python
#convert the text data to feature
vectorizer = TfidfVectorizer()  
```

```python
vector_feature = vectorizer.fit_transform(combine)
print(vector_feature)
```

Cosine Similarity

```python
#get the similarity score using cosine similarity
similarity = cosine_similarity(vector_feature)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#get the movie name from the user
movie_name = input("Enter the movie name:")
```

```python
#creat a list with all the movie names of the data
the_titles = movies['title'].tolist()
print(the_titles) 
```

```python
#find the close match for the movie name by user given
find_match = difflib.get_close_matches(movie_name, the_titles)
print(find_match)
```

```python
close_match = find_match[0]
print(close_match)
```

```python
index_title = movies[movies.title == close_match]['index'].values[0]
print(index_title)
```

```python
#get list of similar index
similar_score_index = list(enumerate(similarity[index_title]))
print(similar_score_index)
```

```python
#sorting movies by similar index
sorted_similar_score_index = sorted(similar_score_index, key=lambda x:x[1], reverse=True) 
print(sorted_similar_score_index)
```

```python
#print the suggest movie use sorted index

print('Movie suggested for you: \n')

i = 1

for movie in sorted_similar_score_index:
    index = movie[0]
    title_by_index = movies[movies.index==index]['title'].values[0]
    if (i<40):
        print(i,'.' , title_by_index)
        i+=1
```

Movie Recommendation System

```python
movie_name = input("Enter the movie name:")
print("Your Search: ",movie_name)
find_match = difflib.get_close_matches(movie_name, the_titles)
close_match = find_match[0]
index_title = movies[movies.title == close_match]['index'].values[0]
similar_score_index = list(enumerate(similarity[index_title]))
sorted_similar_score_index = sorted(similar_score_index, key=lambda x:x[1], reverse=True) 

print('\nMovie suggested for you: ')

i = 1

for movie in sorted_similar_score_index:
    index = movie[0]
    title_by_index = movies[movies.index==index]['title'].values[0]
    if (i<40):
        print(i,'.' , title_by_index)
        i+=1
```
